```python
import statistics
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_statistics(data1, data2=None):
    results = {}

    try:
        # Mean
        results['mean'] = statistics.mean(data1)
    except statistics.StatisticsError as e:
        results['mean'] = str(e)

    try:
        # Median
        results['median'] = statistics.median(data1)
    except statistics.StatisticsError as e:
        results['median'] = str(e)

    try:
        # Mode
        results['mode'] = statistics.mode(data1)
    except statistics.StatisticsError as e:
        results['mode'] = str(e)

    try:
        # Standard Deviation
        results['std_dev'] = statistics.stdev(data1)
    except statistics.StatisticsError as e:
        results['std_dev'] = str(e)

    try:
        # Variance
        results['variance'] = statistics.variance(data1)
    except statistics.StatisticsError as e:
        results['variance'] = str(e)

    if data2:
        try:
            # Percentage Error = (|approximate - exact| / exact) * 100
            percentage_errors = [abs((a - e) / e) * 100 for a, e in zip(data1, data2)]
            results['percentage_error'] = statistics.mean(percentage_errors)
        except ZeroDivisionError as e:
            results['percentage_error'] = str(e)

        try:
            # Percent Difference = (|value1 - value2| / ((value1 + value2)/2)) * 100
            percent_differences = [abs(a - b) / ((a + b) / 2) * 100 for a, b in zip(data1, data2)]
            results['percent_difference'] = statistics.mean(percent_differences)
        except ZeroDivisionError as e:
            results['percent_difference'] = str(e)

        try:
            # Correlation
            results['correlation'] = stats.pearsonr(data1, data2)[0]  # Pearson correlation coefficient
        except Exception as e:
            results['correlation'] = str(e)

        try:
            # Significant Difference (using t-test)
            t_stat, p_value = stats.ttest_ind(data1, data2)
            results['significant_difference'] = p_value < 0.05  # True if p-value < 0.05
        except Exception as e:
            results['significant_difference'] = str(e)

    return results

def plot_data(data1, data2=None):
    plt.figure(figsize=(10, 5))

    # Plotting histogram for data1
    plt.subplot(1, 2, 1)
    plt.hist(data1, bins=10, alpha=0.7, label='Data1')
    plt.title('Histogram of Data1')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

    if data2:
        # Plotting scatter plot if data2 is provided
        plt.subplot(1, 2, 2)
        plt.scatter(data1, data2, alpha=0.7, label='Data1 vs Data2')
        plt.title('Scatter Plot of Data1 vs Data2')
        plt.xlabel('Data1')
        plt.ylabel('Data2')
        plt.legend()

    plt.tight_layout()
    plt.show()

# Example usage
data1 = [10, 20, 20, 30, 40, 50]
data2 = [12, 22, 22, 28, 38, 48]

statistics_results = calculate_statistics(data1, data2)
for key, value in statistics_results.items():
    print(f"{key}: {value}")

plot_data(data1, data2)
```
